```python
from sklearn.naive_bayes import MultinomialNB
import numpy as np

model = MultinomialNB()

def train_model(moves):
    if len(moves) < 2:  # need at least 2 moves
        return None
    X = np.array(moves[:-1]).reshape(-1,1)   # previous moves
    y = np.array(moves[1:])                  # next moves
    model.fit(X,y)
    return model
```

```python
import random

def computer_choice(moves):
    if len(moves) < 5:  # early rounds → random
        return random.choice([0,1,2])
    trained = train_model(moves)
    if trained:
        predicted = model.predict(np.array([moves[-1]]).reshape(-1,1))[0]
    else:
        predicted = random.choice([0,1,2])
    # counter move
    return (predicted + 1) % 3
```

```python
def get_winner(user, computer):
    if user == computer:
        return "It's a tie!"
    elif (user == 0 and computer == 2) or \
         (user == 1 and computer == 0) or \
         (user == 2 and computer == 1):
        return "You win!"
    else:
        return "Computer wins!"
```
